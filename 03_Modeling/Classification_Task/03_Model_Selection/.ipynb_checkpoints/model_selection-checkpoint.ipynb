Arbol de decisión

```python
df = df[df["estado_next"] != "No aplica"]
df = df[df["estado_next"] != "Sin bolsa de creditos"]
df = df[df["estado_next"] != "Graduado"]
df = df[df["estado_next"] != "Pérdida del Beneficio"]
df = df[df["estado_next"] != "Abandono"]
df = df[df["estado_next"].notna()]


df["N_Periodos_adicionales"] = (
    df["periodo_orden"] - df["PERIODOS_BD_SNIES"]
).clip(lower=0)


# Variables predictoras y objetivo
X = df[[
    "N_Aplazado",
    "N_Matriculado",
    "N_Abandono",
    "N_Sin_bolsa_de_creditos",
    "N_Pérdida_del_beneficio",
    "N_Matriculas_adicionales",
    "N_Graduado",
    "pct_aprob_acum",
    "pct_perd_acum",
    "periodo_orden",
    "N_Periodos_adicionales"
]]
y = df["estado_next"]

# Codificar variable objetivo (Estado)
le = LabelEncoder()
y_encoded = le.fit_transform(y)  # convierte "Matriculado", "Graduado", etc. a números

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Modelo de clasificación (Random Forest)
clf = RandomForestClassifier(n_estimators=200, random_state=42, class_weight="balanced")
clf.fit(X_train, y_train)

# Predicciones
y_pred = clf.predict(X_test)

# Evaluación
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nMatriz de confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred, target_names=le.classes_))





def predecir_estado(nuevo, clf, le):
    resultados = []

    for _, row in nuevo.iterrows():
        # ====== REGLAS ======
        if row["N_Aplazado"] >= 4:
            resultados.append("Abandono")

        elif row["N_Abandono"] >= 1:
            resultados.append("Abandono")

        elif row["N_Sin_bolsa_de_creditos"] >= 1:
            resultados.append("Sin_bolsa_de_creditos")

        elif (row["pct_aprob_acum"] + row["pct_perd_acum"] >= 1.1) and \
             (row["N_Periodos_adicionales"] <= 4) and \
             (row["pct_perd_acum"] > 0.1):
            resultados.append("Sin_bolsa_de_creditos")

        elif row["N_Periodos_adicionales"] > 4:
            resultados.append("Pérdida_del_beneficio")

        elif row["N_Pérdida_del_beneficio"] >= 1:
            resultados.append("Pérdida_del_beneficio")

        elif row["N_Graduado"] >= 1:
            resultados.append("Graduado")

        elif row["pct_aprob_acum"] >= 1.0:   # Graduado cuando llega a 100%
            resultados.append("Graduado")

        else:
            # ====== MODELO ======
            X_input = pd.DataFrame([row[clf.feature_names_in_].to_dict()])
            pred = clf.predict(X_input)
            resultados.append(le.inverse_transform(pred)[0])

    return resultados
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ==============================
# División train-test
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# ==============================
# Aplicar SMOTE SOLO al train
# ==============================
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Antes de SMOTE:", dict(zip(*np.unique(y_train, return_counts=True))))
print("Después de SMOTE:", dict(zip(*np.unique(y_train_res, return_counts=True))))

# ==============================
# Entrenar modelo balanceado
# ==============================
clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X_train_res, y_train_res)

# ==============================
# Evaluación
# ==============================
y_pred = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nMatriz de confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred, target_names=le.classes_))
```

Probar varios modelos para elgir el mejor

```python
try:
    from xgboost import XGBClassifier
    XGB_OK = True
except Exception:
    XGB_OK = False


# ===== 1) Configuración de columnas =====
FEATURES = [
    "N_Aplazado",
    "N_Matriculado",
    "N_Abandono",
    "N_Sin_bolsa_de_creditos",
    "N_Pérdida_del_beneficio",
    "N_Matriculas_adicionales",
    "N_Graduado",
    "pct_aprob_acum",
    "pct_perd_acum"
]
TARGET = "estado_next"
GROUP = "DOCUMENTO"

CLASES_EXCLUIR = [
    "No aplica",
    "Sin bolsa de creditos",
    "Graduado",
    "Pérdida del Beneficio"
]

# ===== 2) Limpieza y filtrado =====
def preparar_base(df):
    # Excluir clases indicadas y NaN
    mask = (~df[TARGET].isin(CLASES_EXCLUIR)) & (df[TARGET].notna())
    base = df.loc[mask].copy()

    # Reemplazar inf por NaN y luego imputar en pipeline (no aquí)
    base.replace([np.inf, -np.inf], np.nan, inplace=True)

    # Asegurar que las features existan
    faltantes = [c for c in FEATURES if c not in base.columns]
    if faltantes:
        raise KeyError(f"Faltan columnas en df: {faltantes}")

    # y codificada
    le = LabelEncoder()
    y = le.fit_transform(base[TARGET])

    X = base[FEATURES].copy()
    grupos = base[GROUP].values

    return base, X, y, grupos, le


# ===== 3) Pipelines por modelo =====
def pipe_arbol():
    # Árbol/Forest/GB no requieren escalado; solo imputar
    return Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", DecisionTreeClassifier())
    ])

def pipe_forest():
    return Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", RandomForestClassifier(random_state=42))
    ])

def pipe_gb():
    return Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", GradientBoostingClassifier(random_state=42))
    ])

def pipe_lr():
    # LR sí se beneficia de escalado
    return Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler(with_mean=False)),  # robusto a columnas constantes
        ("model", LogisticRegression(max_iter=2000, n_jobs=None))
    ])

def pipe_xgb():
    return Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", XGBClassifier(
            objective="multi:softprob",
            tree_method="hist",
            random_state=42,
            n_jobs=-1,
            eval_metric="mlogloss"
        ))
    ])


# ===== 4) Grillas =====
param_grids = {
    "DecisionTree": {
        "pipeline": pipe_arbol(),
        "params": {
            "model__max_depth": [3, 5, 8, None],
            "model__min_samples_split": [2, 10, 50],
            "model__min_samples_leaf": [1, 5, 10],
            "model__class_weight": [None, "balanced"]
        }
    },
    "RandomForest": {
        "pipeline": pipe_forest(),
        "params": {
            "model__n_estimators": [200, 500],
            "model__max_depth": [None, 8, 16],
            "model__min_samples_leaf": [1, 5, 10],
            "model__class_weight": [None, "balanced_subsample"]
        }
    },
    "GradientBoosting": {
        "pipeline": pipe_gb(),
        "params": {
            "model__n_estimators": [200, 500],
            "model__learning_rate": [0.05, 0.1],
            "model__max_depth": [2, 3]
        }
    },
    "LogisticRegression": {
        "pipeline": pipe_lr(),
        "params": {
            "model__C": [0.5, 1.0, 2.0, 5.0],
            "model__penalty": ["l2"],
            "model__solver": ["lbfgs"],
            "model__class_weight": [None, "balanced"]
        }
    }
}

if XGB_OK:
    param_grids["XGBoost"] = {
        "pipeline": pipe_xgb(),
        "params": {
            "model__n_estimators": [300, 600],
            "model__max_depth": [3, 6],
            "model__learning_rate": [0.05, 0.1],
            "model__subsample": [0.8, 1.0],
            "model__colsample_bytree": [0.8, 1.0],
        }
    }


# ===== 5) Búsqueda en grilla por grupos =====
def grid_search_panel(df, save_best_as=None):
    base, X, y, grupos, le = preparar_base(df)

    # Split por grupo: 80% train/val, 20% test
    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    train_idx, test_idx = next(gss.split(X, y, groups=grupos))

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    grupos_train = grupos[train_idx]

    # CV interno por grupos
    inner_cv = GroupKFold(n_splits=5)

    resultados = []
    best_overall = None
    best_score = -np.inf

    scoring = {
        "accuracy": "accuracy",
        "balanced_accuracy": "balanced_accuracy",
        "f1_macro": "f1_macro"
    }

    for nombre, cfg in param_grids.items():
        pipe = cfg["pipeline"]
        grid = cfg["params"]

        gs = GridSearchCV(
            estimator=pipe,
            param_grid=grid,
            scoring=scoring,
            refit="balanced_accuracy",   # métrica principal
            cv=inner_cv,
            n_jobs=-1,
            verbose=0
        )

        # Importante: pasar grupos en fit
        gs.fit(X_train, y_train, groups=grupos_train)

        # Evaluación en test (no usado en CV)
        y_pred_test = gs.best_estimator_.predict(X_test)
        acc = accuracy_score(y_test, y_pred_test)
        bacc = balanced_accuracy_score(y_test, y_pred_test)
        f1m = f1_score(y_test, y_pred_test, average="macro")

        result = {
            "modelo": nombre,
            "mejores_params": gs.best_params_,
            "cv_balanced_accuracy": gs.best_score_,
            "test_accuracy": acc,
            "test_balanced_accuracy": bacc,
            "test_f1_macro": f1m,
            "best_estimator": gs.best_estimator_,
            "label_encoder": le
        }
        resultados.append(result)

        if bacc > best_score:
            best_score = bacc
            best_overall = result

        # (Opcional) imprimir resumen rápido
        print(f"\n=== {nombre} ===")
        print("Mejores params:", gs.best_params_)
        print(f"CV balanced_acc: {gs.best_score_:.4f}")
        print(f"TEST  acc/bacc/f1m: {acc:.4f} / {bacc:.4f} / {f1m:.4f}")

    # Reporte extendido del mejor modelo en test
    print("\n==============================")
    print("MEJOR MODELO (por balanced_accuracy en test):", best_overall["modelo"])
    print("Parámetros:", best_overall["mejores_params"])
    print(f"Test balanced_accuracy: {best_overall['test_balanced_accuracy']:.4f}")
    print("\nMatriz de confusión (test):")
    y_test_pred_best = best_overall["best_estimator"].predict(X_test)
    print(confusion_matrix(y_test, y_test_pred_best))
    print("\nReporte de clasificación (test):")
    clases = best_overall["label_encoder"].classes_
    print(classification_report(y_test, y_test_pred_best, target_names=clases))

    # Guardar si se pide
    if save_best_as:
        joblib.dump(best_overall["best_estimator"], f"{save_best_as}_model.pkl")
        joblib.dump(best_overall["label_encoder"], f"{save_best_as}_labelencoder.pkl")
        print(f"\nGuardado: {save_best_as}_model.pkl y {save_best_as}_labelencoder.pkl")

    # Devolver tabla de resultados y mejor
    res_df = pd.DataFrame([
        {
            "modelo": r["modelo"],
            "cv_balanced_accuracy": r["cv_balanced_accuracy"],
            "test_accuracy": r["test_accuracy"],
            "test_balanced_accuracy": r["test_balanced_accuracy"],
            "test_f1_macro": r["test_f1_macro"],
            "mejores_params": r["mejores_params"]
        }
        for r in resultados
    ]).sort_values("test_balanced_accuracy", ascending=False)

    return best_overall, res_df


# ===== 6) EJECUCIÓN =====
best, tabla_resultados = grid_search_panel(df, save_best_as="mejor_modelo_panel")

# Ver tabla comparativa
print("\nTabla comparativa de modelos:")
print(tabla_resultados.to_string(index=False))
```
