### **Graduados y terminación de materias**

```python
df_graduados = utils.calcular_estado_por_periodo(data_filtered, estado="GRADUADO")
utils.plot_estudiantes_por_periodo(df_graduados, estado="GRADUADO")

df_terminacion_materias = utils.calcular_estado_por_periodo(data_filtered, estado="TERMINACION DE MATERIAS")
utils.plot_estudiantes_por_periodo(df_terminacion_materias, estado="TERMINACION DE MATERIAS")

tabla_graduacion = pd.merge(df_terminacion_materias,df_graduados, on ="Periodo")

tabla_graduacion['TOTAL'] = tabla_graduacion['TERMINACION DE MATERIAS'] + tabla_graduacion['GRADUADO']
utils.plot_estudiantes_por_periodo(tabla_graduacion, estado="TOTAL")
```
